```python
import pandas as pd
import numpy as np
import random 
```

```python
amostra=[random.randint(1,99) for _ in range(20)]
dados=np.array(amostra)
print(dados)
```

```python
#Calculo dos Quartis:
q1=np.percentile(dados,25) # o primeiro quartil representa 25% do total de amostra.
q2=np.percentile(dados,50) # mediana pode ser interpretada como o seu segundo quartil (quartil 50)
q3=np.percentile(dados,75) # o terceiro quartil repersenta do total da amostra

#Exibição:
print(f"primeiro Quartil(Q1): {q1}")
print(f"Segundo Quartil(Q2):{q2}")
print(f"Terceiro Quartil (Q3): {q3}")
```
